```python
from scipy.stats import mstats

#for a single well, find the slope of best fit
def WellFit(well,time_all,theil):
    #if theil then use median based fitting, else use least squares.
    valid_points = (well > 0) #ignore invalid values
    
    if theil:
        if (np.sum(valid_points) > 2): #do we have enough points to attempt fitting?
            #note that scipy uses masked arrays, so in case of NaNs in future, might be good to rewrite
            a, b, a_low, a_high = mstats.theilslopes(np.log(well[valid_points]),x=time_all[valid_points],alpha=0.682)
            a_v = max(abs(a - a_low),abs(a - a_high))**2
        else:
            #bad style to return NaNs when using scipy elsewhere
            a = np.NaN
            a_low = np.NaN
            a_high = np.NaN
            a_v = np.NaN
        
        #returning a_v for compatibility with old code, but it is better to use a_low, a_high
        return a, a_v, a_low, a_high, b
    else:
        if (np.sum(valid_points) > 2): #do we have enough points to attempt fitting?
            [a,_], [[av,_],[_,_]] = np.polyfit(time_all[valid_points],np.log(well[valid_points]),1,full=False,cov=True)
        else:
            a = np.NaN
            av = np.NaN
        
        return a, av

#find fits for all wells
def FindFits(C1_all,C2_all,time_all = np.array([]),theil=False):
    fitC1 = np.zeros((C1_all.shape[1],C1_all.shape[2]), dtype=np.dtype(float))
    fitC2 = np.zeros((C2_all.shape[1],C2_all.shape[2]), dtype=np.dtype(float))
    
    fitC1_v = np.zeros((C1_all.shape[1],C1_all.shape[2]), dtype=np.dtype(float))
    fitC2_v = np.zeros((C2_all.shape[1],C2_all.shape[2]), dtype=np.dtype(float))
    
    fitC1_b = np.zeros((C1_all.shape[1],C1_all.shape[2]), dtype=np.dtype(float))
    fitC2_b = np.zeros((C2_all.shape[1],C2_all.shape[2]), dtype=np.dtype(float))
    
    #if not time_all was passed
    if time_all.size == 0:
        time_all = np.array(range(C1_all.shape[0]))
    
    for i in range(C1_all.shape[1]):
        for j in range(C1_all.shape[2]):
            #lets get the well
            C1_well = C1_all[:,i,j]
            C2_well = C2_all[:,i,j]
            
            #now get the fits
            if theil:
                fitC1[i,j], fitC1_v[i,j],_,_,fitC1_b[i,j] = WellFit(C1_well,time_all,theil)
                fitC2[i,j], fitC2_v[i,j],_,_,fitC2_b[i,j] = WellFit(C2_well,time_all,theil)
            else:
                fitC1[i,j], fitC1_v[i,j] = WellFit(C1_well,time_all,theil)
                fitC2[i,j], fitC2_v[i,j] = WellFit(C2_well,time_all,theil)
    
    return fitC1, fitC2, fitC1_v, fitC2_v, fitC1_b, fitC2_b

#use this to focus on particular parts of growth phase.
t_min = 0
t_max = 30

#notice the backwards here, this is on purpose to account for the labeling in Plate5.
#See discussion in ALK_EGT on June 22nd, 2016

fitR, fitG, fitR_v, fitG_v, fitR_b, fitG_b = FindFits(glanp_t[t_min:t_max,:,:],rlanp_t[t_min:t_max,:,:],theil=True)
```

```python
#some code to visualize individual wells

from ipywidgets import interact
import ipywidgets as widgets

matplotlib.rcParams['figure.figsize'] = (6, 6)

def PlotWell(x,y,y_low = 2000, y_high = 750000, \
             use_log = True, use_legend = True, use_save = False, use_prop = True, use_grid = False, \
             return_ax = False, use_labels = True):
    #x,y are the position on the plate, y_low, y_high are the scale for the axis with defaults
    time_all = np.array(range(t_max))
    
    #note that rlanp_t is channel 1, which in this case is resistant and should be plotted in green
    plt.plot(4*time_all,rlanp_t[:t_max,x,y],g_GC+'o',fillstyle='none', label="Resistant (C1)")
    
    #get the fit parameters
    a_G = fitG[x,y]
    b_G = fitG_b[x,y]
    
    fit_G = np.exp(time_all*a_G + b_G)
    plt.plot(4*time_all,fit_G,g_GC+'-')
    
    plt.plot(4*time_all,glanp_t[:t_max,x,y],r_GC+'o',fillstyle='none', label="Parental (C2)")
    
    #get the fit parameters
    a_R = fitR[x,y]
    b_R = fitR_b[x,y]
    
    fit_R = np.exp(time_all*a_R + b_R)
    plt.plot(4*time_all,fit_R,r_GC+'-')
    
    plt_title = 'Well ' + p_let[x] + p_num[y] + ': '
    
    if drugV[x,y]:
        plt_title = plt_title + 'Alectinib'
    else:
        plt_title = plt_title + ' DMSO'
    
    if fibroV[x,y]:
        plt_title = plt_title + ' + CAF'
    else:
        plt_title = plt_title + ''
    
    if use_prop:
        plt_title = plt_title + ' (p(0) = ' + str(propP[x,y]) + ')'
    
    if use_labels:
        plt.title(plt_title)
        
        plt.xlabel('Time (h)')
        plt.ylabel('Fluorescent area')
    
    if use_legend:
        plt.legend(loc = 'lower right')
    
    if use_log:
        plt.yscale('log')
    
    plt.ylim([y_low,y_high])
    plt.grid(use_grid, which='both')
    
    if use_save:
        plt.savefig('Well_' + p_let[x] + p_num[y] + '_tS.png',dpi=500)
    
    if return_ax:
        return plt.gca()
    else:
        plt.show()
    
    if use_legend:
        print('Parental growth rate: ' + str(fitR[x,y]))
        print('Resistant growth rate: ' + str(fitG[x,y]))

#create an interact that lets us navigate the wells
interact(PlotWell, x = widgets.IntSlider(min = 0, max = (rlanp_t.shape[1] - 1), step = 1,value = 0),
         y = widgets.IntSlider(min = 0, max = (rlanp_t.shape[2] - 1), step=1, value=0))
```

```python
#let's look at how the imaging proportions and densities compare to the desired seeding proportions and densities

matplotlib.rcParams['figure.figsize'] = (8, 8)

#auxilary function to get mask of points with desired seeding densities
def getValidPoints(well_guide,threshold):
    valid_points = (((well_guide >= threshold) & (well_guide <= (1 - threshold))) | (well_guide == 0) | (well_guide == 1))
    return valid_points

def fitplot(x,y,c):
    plt.plot(x,y,c+'o',fillstyle='none')
    
    #get the line of best fit
    [a_fit,b_fit] = np.polyfit(x,y,1)
    
    #plot the line of best fit
    x_min = np.amin(x)
    x_max = np.amax(x)
    plt.plot([x_min,x_max],[a_fit*x_min + b_fit,a_fit*x_max + b_fit],c+'-')
    
    return a_fit, b_fit

def ActualPropPlot(C1_slice,C2_slice,well_guide,threshold = 0):
    propFA = C2_slice/(C1_slice + C2_slice)
    
    valid_points = getValidPoints(well_guide,threshold)
    
    #the well_guide <= 1 needed to handle NaNs
    #will add line of bestfit (least squares) as solid
    fitplot(well_guide[valid_points].flatten(),propFA[valid_points].flatten(),'k')
    
    plt.plot([0,1],[0,1],'k--')
    plt.xlabel('Desired seeding proportion')
    plt.ylabel('Actual fluorescent area proportion in first image')
    plt.xlim([0,1])
    plt.ylim([0,1])
    
    plt.show()

def ActualDensityPlot(C1_slice,C2_slice,well_guide,ymax = 400000,threshold = 0):
    densityFA = C1_slice + C2_slice
    
    mean_density = np.mean(densityFA)
    
    valid_points = getValidPoints(well_guide,threshold)
    
    #(the well_guide <= 1 needed to handle NaNs)
    fitplot(well_guide[valid_points].flatten(),C1_slice[valid_points].flatten(),g_GC)
    fitplot(well_guide[valid_points].flatten(),C2_slice[valid_points].flatten(),r_GC)
    
    print("Mean density is: " + str(mean_density))
    
    plt.plot(well_guide.flatten(),densityFA.flatten(),'ko')
    plt.plot([0,1],[mean_density, mean_density], 'k--')
    plt.ylim([0,ymax])
    
    plt.show()
    
ActualPropPlot(rlanp_t[0,:,:],glanp_t[0,:,:],propP)
ActualDensityPlot(rlanp_t[0,:,:],glanp_t[0,:,:],propP)
```

```python
#this renormalizes the units to range from zero to seed_density by using best fit line
def NormalizeFAUnitsByFit(data,well_guide,seed_density,b_sub, threshold, hard_floor):
    data_zero = data[0,:,:]
    
    valid_points = getValidPoints(well_guide,threshold)
    
    #fit the data
    [a_fit,b_fit], [[av,abv],[_,bv]] = np.polyfit(well_guide[valid_points].flatten(),data_zero[valid_points].flatten(),1,full=False,cov=True)
    
    if b_sub:
        out_data = (data - b_fit)*(seed_density/a_fit)
        out_data_v = (data*data*(av/a_fit**2) + av*(b_fit**2/a_fit**2) - 2*(b_fit/a_fit)*abv + bv)*(seed_density**2/a_fit**2)
    else:
        out_data = data*(seed_density/(a_fit + b_fit))
        out_data_v = data*data*((seed_density/(a_fit + b_fit))**2)*((av + bv + 2*abv)/(a_fit + b_fit)**2)
    
    if hard_floor:
        out_data = np.maximum(out_data,0)
    
    return out_data, out_data_v

def AlignFAUnitsByFit(C1_all,C2_all,well_guide,background_offset = True,threshold = 0, hard_floor = False):
    #calculate seeding density
    meanFAdensity = np.mean(rlanp_t[0,:,:] + glanp_t[0,:,:])
    
    #Normalize each channel
    cleanC1_all, cleanC1_all_v = NormalizeFAUnitsByFit(C1_all,1 - well_guide,meanFAdensity,background_offset,threshold,hard_floor)
    cleanC2_all, cleanC2_all_v = NormalizeFAUnitsByFit(C2_all,well_guide,meanFAdensity,background_offset,threshold,hard_floor)
    
    return cleanC1_all, cleanC2_all, cleanC1_all_v, cleanC2_all_v

def ProportionsWithError(C1,C2,C1_v,C2_v):
    props = C2/(C1 + C2)
    props_v = props*props*(C2_v/(C2*C2) + (C1_v + C2_v)/((C1 + C2)**2))
    
    bad_values = ((C1 < 0)|(C2 < 0)|(C1 + C2 == 0))
    props[bad_values] = np.NaN
    props_v[bad_values] = np.NaN
    
    return props, props_v

#adjust the units
cC1all, cC2all, cC1all_v, cC2all_v = AlignFAUnitsByFit(rlanp_t,glanp_t,propP,background_offset = True,threshold = 0.1, hard_floor = True)

#propFA is used by later processing code
propFA, propFA_v = ProportionsWithError(cC1all, cC2all,cC1all_v,cC2all_v)

#take a look
ActualPropPlot(cC1all[0,:,:],cC2all[0,:,:],propP,threshold = 0.1)
ActualDensityPlot(cC1all[0,:,:],cC2all[0,:,:],propP,threshold = 0.1)
```

```python
import os
import time
import FluorescentArea as fa

def LoadPlateTimeSeries(rootDir, subDirMatch, col_let, row_num):
    start_tot = time.clock()
    
    time_all = []
    imgP_all = []
    imgC1_all = []
    imgC2_all = []
    imgFA1_all = []
    imgFA2_all = []
    
    time_step = -1
    
    for dirName, subdirList, fileList in os.walk(rootDir):
        if subDirMatch in subdirList:
            #track time
            time_step = time_step + 1
            start_time = time.clock()
            
            time_all.append(time.strptime(dirName[len(rootDir):],'/%y%m/%d/%H%M'))
            
            imgP, imgC1, imgC2 = fa.FullLoad(col_let,row_num,1,dirName + '/' + subDirMatch +'/')
            
            imgP_all.append(imgP)
            imgC1_all.append(imgC1)
            imgC2_all.append(imgC2)
            
            #Process Images as in Original
            faC1 = fa.FluorescentAreaMark(imgC1)
            #for channel 2 use wider grid because of big saturation blocks and little to no vignetting
            faC2 = fa.FluorescentAreaMark(imgC2, gridSize = 5) 
            
            imgFA1_all.append(faC1)
            imgFA2_all.append(faC2)
            
            end_time = time.clock()
            
            print(str(time_step) + ': ' + dirName + ' took ' + str(end_time - start_time) + ' seconds')
    
    end_tot = time.clock()
    print('Total time taken: ' + str(end_tot - start_tot) + ' seconds')
    
    return imgP_all, imgC1_all, imgC2_all, imgFA1_all, imgFA2_all, time_all
```

```python
minR = np.amin(fitR)
maxR = np.amax(fitR)
print([minR, maxR])

minG = np.amin(fitG)
maxG = np.amax(fitG)
print([minG, maxG])

#y_min = min(minR,minG)
#y_max = max(maxR,maxG)

y_min = -0.03
y_max = 0.06
ms = 5

rp = True #should we use image props or seed props?
txe = False #should we propogate error in image props?

pv = [0.1,0.2,0.4,0.6,0.8,0.9]
cl = 3 #from which time point do we get seeding proportions?

matplotlib.rcParams['figure.figsize'] = (16, 4)
matplotlib.rcParams['figure.dpi'] = 500

def GetFFData(ind_cond,ends,prop_vals,real_prop,track_xerr):
    #put all the data into x,yR,yR_e,yG,yG_e
    x = np.array([])
    x_e = np.array([])
    yR = np.array([])
    yR_e = np.array([])
    yG = np.array([])
    yG_e = np.array([])
    
    #use the to store the gain functions
    yGain = np.array([])
    yGain_e = np.array([])
    
    for prop_val in prop_vals:
        ind_spec = np.where((propP == prop_val)*ind_cond)
        
        for (i,j) in zip(ind_spec[0],ind_spec[1]):
            if real_prop:
                if (propFA[cl,i,j] <= 1): #ignore the NaNds
                    x = np.append(x,propFA[cl,i,j])
                    if track_xerr:
                        x_e = np.append(x_e,np.sqrt(propFA_v[cl,i,j]))
                    else:
                        x_e = np.append(x_e,0)
            else:
                x = np.append(x,prop_val)
                x_e = np.append(x_e,0)
            
            if ((propFA[cl,i,j] <= 1) | (not real_prop)):
                yR = np.append(yR,fitR[i,j])
                yR_e = np.append(yR_e,np.sqrt(fitR_v[i,j]))
                yG = np.append(yG,fitG[i,j])
                yG_e = np.append(yG_e,np.sqrt(fitG_v[i,j]))
        
    yR_body = yR
    yG_body = yG
    
    if ends:
        #add the parental monocultures
        xR = x
        xR_e = x_e
        ind_spec = np.where((propP == 1)*ind_cond)
        for (i,j) in zip(ind_spec[0],ind_spec[1]):
            xR = np.append(xR,1)
            xR_e = np.append(xR_e,0)
            yR = np.append(yR,fitR[i,j])
            yR_e = np.append(yR_e,np.sqrt(fitR_v[i,j]))
        
        xG = x
        xG_e = x_e
        ind_spec = np.where((propP == 0)*ind_cond)
        for (i,j) in zip(ind_spec[0],ind_spec[1]):
            xG = np.append(xG,0)
            xG_e = np.append(xG_e,0)
            yG = np.append(yG,fitG[i,j])
            yG_e = np.append(yG_e,np.sqrt(fitG_v[i,j]))
    else:
        xR = x
        xR_e = x_e
        xG = x
        xG_e = x_e
    
    return xR,xR_e,yR,yR_e,xG,xG_e,yG,yG_e

def FFPlot(ind_cond,ends=False,gains=False, prop_vals = [0.025,0.05,0.1,0.2,0.4,0.6,0.8,0.9,0.95,0.975], real_prop = False, track_xerr = True, direct_gains = False):
    xR,xR_e,yR,yR_e,xG,xG_e,yG,yG_e = GetFFData(ind_cond,ends,prop_vals,real_prop,track_xerr)
    
    #plot the data
    if real_prop:
        plt.errorbar(xR,yR,xerr=xR_e,yerr=yR_e,fmt=r_GC+'s',markersize=ms,fillstyle='none')
        plt.errorbar(xG,yG,xerr=xG_e,yerr=yG_e,fmt=g_GC+'s',markersize=ms,fillstyle='none')
    else:
        plt.errorbar(xR,yR,yerr=yR_e,fmt=r_GC+'s',markersize=ms,fillstyle='none')
        plt.errorbar(xG,yG,yerr=yG_e,fmt=g_GC+'s',markersize=ms,fillstyle='none')
    
    #plot the x-axis
    plt.plot([0, 1],[0, 0],'k-')
    
    #find the line of best fit
    if real_prop:
        [ra,rb], [[rav,rabv],[_,rbv]] = np.polyfit(xR,yR,1,w=1/np.sqrt(yR_e**2 + xR_e**2),full=False,cov=True)
        
        [ga,gb], [[gav,gabv],[_,gbv]] = np.polyfit(xG,yG,1,w=1/np.sqrt(yG_e**2 + xG_e**2),full=False,cov=True)
    else:
        [ra,rb], [[rav,rabv],[_,rbv]] = np.polyfit(xR,yR,1,w=1/yR_e**2,full=False,cov=True)
        [ga,gb], [[gav,gabv],[_,gbv]] = np.polyfit(xG,yG,1,w=1/yG_e**2,full=False,cov=True)
    
    #plot the fits
    plt.plot([0,1],[rb,rb + ra],r_GC+'--')
    plt.plot([0,1],[gb,gb + ga],g_GC+'--')
    
    if gains:
        gain_a = ra - ga
        gain_av = rav + gav
        
        gain_b = rb - gb
        gain_bv = rbv + gbv
        
        gain_abv = rabv + gabv
        
        #plot the gain function
        gain, = plt.plot([0,1],[gain_b, gain_b + gain_a],'k-.', label='Gain Fn.')
        
        #the crossing point
        gain_p = -gain_b/gain_a
        gain_pv = gain_p**2*(gain_av/gain_a**2 + gain_bv/gain_b**2 - 2*gain_abv/(gain_a*gain_b))
        
        #visualize the crossing point error
        if gain_p < 1 and gain_p > 0:
            plt.fill_between([gain_p - np.sqrt(gain_pv),gain_p,gain_p + np.sqrt(gain_pv)],[-gain_a*np.sqrt(gain_pv),0, gain_a*np.sqrt(gain_pv)], color='silver')
            print('Has a ' + ('' if (a < 0) else 'un') + 'stable fixed point at ' + str(gain_p) + 
                  ' of width +-' + str(np.sqrt(gain_pv)) + 
                  ' and height +-' + str(np.abs(gain_a*np.sqrt(gain_pv))))
    else:
        gain = []
        gain_p = np.NaN
        gain_pv = np.NaN
        gain_a = np.NaN
        gain_av = np.NaN
    
    plt.xlim([0,1])
    plt.xlabel('Initial proportion parental')
    plt.ylim([y_min,y_max])
    #plt.ylabel('Growth rate')
    plt.grid(True)
    
    return gain, np.array([gain_p, gain_a, gain_pv, gain_av]), np.array([ra, rb, rav, rbv, rabv]), np.array([ga, gb, gav, gbv, gabv])

def FitRightPoint(fit):
    out = fit[0] + fit[1]
    out_v = fit[2] + fit[3] + 2*fit[4]
    
    return [out, out_v]

def IndepDiff(point1,point2):
    out = point1[0] - point2[0]
    out_v = point1[1] + point2[1]
    
    return [out, out_v]

def GameCoord(fit_L,fit_R):
    point_y = IndepDiff(fit_R[[1,3]],fit_L[[1,3]])
    point_x = IndepDiff(FitRightPoint(fit_L),FitRightPoint(fit_R))
    
    return point_x, point_y

p_all = np.array([])
a_all = np.array([])
p_e_all = np.array([])
a_e_all = np.array([])

game_x_all = np.array([])
game_y_all = np.array([])
game_xe_all = np.array([])
game_ye_all = np.array([]) 
game_labels = np.array([])

#code in case we want the lines later
ra_all = np.array([])
rb_all = np.array([])
ga_all = np.array([])
gb_all = np.array([])

k = 1
game_labels = np.array([])

for drug_flip in [1,0]:
    for fibro_flip in [1,0]:
        #set subplot
        plt.subplot(1,4,k)
        
        #plot the main figure
        gain, [p, a, p_v, a_v], r_fit, g_fit = \
            FFPlot((drug_flip + ((-1)**drug_flip)*drugV)*(fibro_flip + ((-1)**fibro_flip)*fibroV),
                   ends=True,gains=True,real_prop=rp,prop_vals = pv, track_xerr = txe)
        plt.title(('DMSO' if drug_flip else 'Alectinib') + ('' if fibro_flip else ' + CAF'))
        
        #add labels
        if k == 1:
            red = matplotlib.patches.Patch(color=r_GC, label='Parental')
            green = matplotlib.patches.Patch(color=g_GC, label='Resistant')
            
            plt.legend(handles=[red,green,gain], loc = 'lower left')
            plt.ylabel('Growth rate')
        
        #increment subplot count
        k = k + 1
        
        #record info
        print('\n\n' + ('No' if drug_flip else 'Yes') + ' Drug + ' + ('No' if fibro_flip else 'Yes') + ' Fibroblasts')
        
        #print the game matrix [A B; C D]
        print(r_fit[0] + r_fit[1]) #A
        print(r_fit[1]) #B
        print(g_fit[0] + g_fit[1]) #C
        print(g_fit[1]) #D
        
        print([[p, np.sqrt(p_v)],[a,np.sqrt(a_v)]])
        p_all = np.append(p_all,p)
        a_all = np.append(a_all,a)
        p_e_all = np.append(p_e_all,np.sqrt(p_v))
        a_e_all = np.append(a_e_all,np.sqrt(a_v))
        
        [[gx,gxv],[gy,gyv]] = GameCoord(g_fit,r_fit)
        game_x_all = np.append(game_x_all,gx)
        game_y_all = np.append(game_y_all,gy)
        game_xe_all = np.append(game_xe_all,np.sqrt(gxv))
        game_ye_all = np.append(game_ye_all,np.sqrt(gyv))
        #game_labels = np.append(game_labels,('N' if drug_flip else 'Y') + ('N' if fibro_flip else 'Y'))
        game_labels = np.append(game_labels,('DMSO' if drug_flip else 'Alectinib') + ('' if fibro_flip else ' + CAF'))
        
        #print the game coordinates:
        print("\nGame coordinates:")
        print(gx)
        print(gy)
        
        ra_all = np.append(ra_all,r_fit[0])
        rb_all = np.append(rb_all,r_fit[1])
        ga_all = np.append(ga_all,g_fit[0])
        gb_all = np.append(gb_all,g_fit[1])

#plt.suptitle('Fitness functions for competition of parental vs. resistant NSCLC', fontsize=16)

if True:
    plt.savefig('RvP_yerr_autogen.pdf',dpi=300,bbox_inches='tight')

plt.show()
```

```python
import math as math

matplotlib.rcParams['figure.figsize'] = (16, 4)

poly_degree = 6

def FFPlotPoly(p_ind,ind_cond,ends=False,gains=False, prop_vals = [0.025,0.05,0.1,0.2,0.4,0.6,0.8,0.9,0.95,0.975], 
               real_prop = False, track_xerr = True, direct_gains = False, pd_ff = [poly_degree,poly_degree]):
    
    xR,xR_e,yR,yR_e,xG,xG_e,yG,yG_e = GetFFData(ind_cond,ends,prop_vals,real_prop,track_xerr)
    
    #plot the data
    if real_prop:
        plt.errorbar(xR,yR,xerr=xR_e,yerr=yR_e,fmt=r_GC+'s',markersize=ms,fillstyle='none')
        plt.errorbar(xG,yG,xerr=xG_e,yerr=yG_e,fmt=g_GC+'s',markersize=ms,fillstyle='none')
    else:
        plt.errorbar(xR,yR,yerr=yR_e,fmt=r_GS+'s',markersize=ms,fillstyle='none')
        plt.errorbar(xG,yG,yerr=yG_e,fmt=g_GS+'s',markersize=ms,fillstyle='none')
    
    #plot the x-axis
    plt.plot([0, 1],[0, 0],'k-')
    
    #set the error weight functions
    if real_prop:
        wR = 1/np.sqrt(yR_e**2 + xR_e**2)
        wG = 1/np.sqrt(yG_e**2 + xG_e**2)
    else:
        wR = 1/yR_e**2
        wG = 1/yG_e**2
    
    #find the poly of best fit, note that this is controlled by the global variable poly_degree
    r_coefs, rV = np.polyfit(xR,yR,pd_ff[0],w=wR,full=False,cov=True)
    g_coefs, gV = np.polyfit(xG,yG,pd_ff[1],w=wG,full=False,cov=True)
    
    game_x = np.sum(g_coefs) - np.sum(r_coefs)
    game_y = r_coefs[-1] - g_coefs[-1]
    
    print(r_coefs)
    print(g_coefs)
    #plot the fits
    x_vals = np.linspace(0,1,500)
    r_fit = np.polyval(r_coefs,x_vals) #note that polyval and polyfit use opposite coefficient orders
    g_fit = np.polyval(g_coefs,x_vals)
    plt.plot(x_vals,r_fit,r_GC+'--')
    plt.plot(x_vals,g_fit,g_GC+'--')
    
    r_fit_data = np.polyval(r_coefs,xR)
    g_fit_data = np.polyval(g_coefs,xG)
    
    r_resid = wR*(yR - r_fit_data)
    g_resid = wG*(yG - g_fit_data)
    
    r_L = sum(r_resid**2)
    g_L = sum(g_resid**2)
    
    print('red data ponts ' + str(len(xR)) + ' green data points ' + str(len(xG)))
    
    r_AIC = 2*(pd_ff[0] + 2) + len(xR)*math.log(r_L)
    g_AIC = 2*(pd_ff[1] + 2) + len(xG)*math.log(g_L)
    r_BIC = math.log(len(xR))*(pd_ff[0] + 2) + len(xR)*math.log(r_L)
    g_BIC = math.log(len(xG))*(pd_ff[1] + 2) + len(xG)*math.log(g_L)
    
    print('AIC, ' + r_GC + ': ' + str(r_AIC) + ' ' + g_GC + ': ' + str(g_AIC))
    print('BIC, ' + r_GC + ': ' + str(r_BIC) + ' ' + g_GC + ': ' + str(g_BIC))
    
    if gains:
        #plot the gain function
        gain, = plt.plot(x_vals,r_fit - g_fit,'k-.', label='Gain Fn.')
        
        #the crossing points
        gain = r_fit - g_fit
        num_crossings = sum(gain[0:-2]*gain[1:-1] < 0)
        print('\nGain function crosses zero a total of ' + str(num_crossings) + ' times')
        
        if num_crossings > 0:
            print('With the crosses at: ' + str(np.flatnonzero(gain[0:-2]*gain[1:-1] < 0)/500))
            
        fp_support = (np.abs(gain) < p_height[p_ind])
        num_starts = sum(((1 - fp_support[0:-2])*fp_support[1:-1]) > 0)
        num_ends = sum((fp_support[0:-2]*(1 - fp_support[1:-1])) > 0)
        
        if (num_starts + num_ends) > 0:
            print('Fixed point region starts ' + str(num_starts) + ' times and ends ' + str(num_ends) + ' times')
            print('The regions are:')
            
            last_point = 0
            end_num = 0
            
            while end_num < num_ends:
                s_point = last_point + np.flatnonzero(fp_support[last_point:])[0]
                last_point = s_point + np.flatnonzero(fp_support[s_point:] - 1)[0]
                print('  start: ' + str(s_point/500) + ' end: ' + str((last_point - 1)/500))
                end_num = end_num + 1
            
            if (num_starts > num_ends) or ((num_starts == num_ends) and (fp_support[0] > 0)):
                s_point = last_point + np.flatnonzero(fp_support[last_point:])[0]
                print('  start: ' + str(s_point/500) + ' end: 1')
    else:
        gain = []
        gain_p = np.NaN
        gain_pv = np.NaN
        gain_a = np.NaN
        gain_av = np.NaN
    
    plt.xlim([0,1])
    plt.xlabel('Initial proportion parental')
    plt.ylim([y_min,y_max])
    #plt.ylabel('Growth rate')
    plt.grid(True)
    
    return [r_AIC, g_AIC, r_BIC, g_BIC], [game_x, game_y]

k = 1

for drug_flip in [1,0]:
    for fibro_flip in [1,0]:
        #set subplot
        plt.subplot(1,4,k)
        
        #record info
        print('\n\n' + ('No' if drug_flip else 'Yes') + ' Drug + ' + ('No' if fibro_flip else 'Yes') + ' Fibroblasts')
        
        #plot the main figure
        ICvals[poly_degree,k - 1,:], gamePoints[poly_degree,k - 1,:] = \
            FFPlotPoly(k - 1,(drug_flip + ((-1)**drug_flip)*drugV)*(fibro_flip + ((-1)**fibro_flip)*fibroV),
                   ends=True,gains=True,real_prop=rp,prop_vals = pv, track_xerr = txe)
        plt.title(('DMSO' if drug_flip else 'Alectinib') + ('' if fibro_flip else ' + CAF'))
        
        k = k + 1       
        

plt.show()
```

```python
plot_fixed_points = False

def GetPlotStyle(i,j,comp_bins,comp_low):
    if ((np.mean(propFA[comp_low:(comp_low+comp_bins),i,j]) - np.std(propFA[comp_low:(comp_low+comp_bins),i,j])) 
            > (np.mean(propFA[comp_bins:,i,j]) + np.std(propFA[comp_bins:,i,j]))):
        plot_style = g_GC+'.-'
    elif ((np.mean(propFA[comp_low:(comp_low+comp_bins),i,j]) + np.std(propFA[comp_low:(comp_low+comp_bins),i,j])) 
            < (np.mean(propFA[comp_bins:,i,j]) - np.std(propFA[comp_bins:,i,j]))):
        plot_style = r_GC+'.-'
    else:
        plot_style = 'k.-'
    
    return plot_style

def PropPlot(ind_cond,prop_val = [0,0.025,0.05,0.1,0.2,0.4,0.6,0.8,0.9,0.95,0.975,1],comp_bins=5,comp_low=0,ms=6):
    ind_spec = np.where(ind_cond)
    
    #first plot all the black background
    for (i,j) in zip(ind_spec[0],ind_spec[1]):
        if propP[i,j] in pv:
            plot_style = GetPlotStyle(i,j,comp_bins,comp_low)
            
            if plot_style == 'k.-':
                plt.plot(np.arange(t_min,(comp_low + 1)),propFA[t_min:(comp_low+1),i,j],'-',color='0.25',
                         markersize=ms,zorder=0)
                plt.plot(np.arange(comp_low,t_max),propFA[comp_low:t_max,i,j],plot_style,markersize=ms,zorder=0)
    
    #plot the g_NC foreground
    for (i,j) in zip(ind_spec[0],ind_spec[1]):
        if propP[i,j] in pv:
            plot_style = GetPlotStyle(i,j,comp_bins,comp_low)
            
            if plot_style == g_GC+'.-':
                plt.plot(np.arange(t_min,(comp_low + 1)),propFA[t_min:(comp_low+1),i,j],'-',color='0.25',
                         markersize=ms,zorder=1)
                plt.plot(np.arange(comp_low,t_max),propFA[comp_low:t_max,i,j],plot_style,markersize=ms,zorder=1)
    
    #plot the r_NC foreground
    for (i,j) in zip(ind_spec[0],ind_spec[1]):
        if propP[i,j] in pv:
            plot_style = GetPlotStyle(i,j,comp_bins,comp_low)
            
            if plot_style == r_GC+'.-':
                plt.plot(np.arange(t_min,(comp_low + 1)),propFA[t_min:(comp_low+1),i,j],'-',color='0.25',
                         markersize=ms,zorder=2)
                plt.plot(np.arange(comp_low,t_max),propFA[comp_low:t_max,i,j],plot_style,markersize=ms,zorder=2)
    
    plt.ylabel("Proportion of Parental (C2)")
    plt.xlim([0,30])
    plt.ylim([0,1])
    plt.grid(zorder=0)

pv = [0.1,0.2,0.4,0.6,0.8,0.9]
cl = 3

matplotlib.rcParams['figure.figsize'] = (16, 4)

plt.subplot(1,4,1)
PropPlot((1 - drugV)*(1 - fibroV), prop_val = pv, comp_low = cl)
plt.title('DMSO')

plt.subplot(1,4,2)
PropPlot((1 - drugV)*fibroV, prop_val = pv, comp_low = cl)
if plot_fixed_points:
    plt.fill_between([0,30],[p_all[1] + p_e_all[1],p_all[1] + p_e_all[1]],
                 [p_all[1] - p_e_all[1],p_all[1] - p_e_all[1]],color='silver')
plt.title('DMSO + CAF')

plt.subplot(1,4,3)
PropPlot(drugV*(1 -  fibroV), prop_val = pv, comp_low = cl)
plt.title('Alectinib')

plt.subplot(1,4,4)
PropPlot(drugV*fibroV, prop_val = pv, comp_low = cl)
plt.title('Alectinib + CAF')

#plt.suptitle('Proportion of parental cells during competition of parental vs. resistant NSCLC', fontsize=16)

if False:
    plt.savefig('PvT_autogen.png',dpi=500)

plt.show()
```
